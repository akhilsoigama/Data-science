**FREQUENT PATTERN GROWTH ALGORITHM**

```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', 500)
dataset = pd.read_csv('groceries - groceries.csv', sep=';')

dataset = dataset.iloc[:, :11]

dataset.replace('', np.nan, inplace=True)

dataset.head()
```

```python
market = []
for i in range(dataset.shape[0]):
  cus = []
  for j in dataset.columns:
    if type(dataset[j][i]) == str:
      cus.append(dataset[j][i])
  market.append(cus)
```

```python
lst = []
for i in market:
  for j in i:
    lst.append(j)
```

```python
import collections
collect = collections.Counter(lst)
```

```python
d = {'Item Name':collect.keys(), 'values':collect.values()}
df = pd.DataFrame(d).sort_values(by='values', ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
new_df = pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fp = fpgrowth(new_df, min_support=0.01, use_colnames=True,verbose=0).sort_values(by='support')
```
