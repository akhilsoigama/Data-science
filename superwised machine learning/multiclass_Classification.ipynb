```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = sns.load_dataset('iris')

data = data.rename(columns={
    'sepal_length': 'sepal_lenght',
    'sepal_width': 'sepal_widht',
    'petal_length': 'petal_lenght',
    'petal_width': 'petal_width',
    'species': 'spices'
})

if len(data) < 500:
    data = pd.concat([data] * (500 // len(data) + 1), ignore_index=True).head(500)

data.head()
```

```python
data['spices'].unique()
```

```python
sns.pairplot(data, hue='spices')
plt.show()
```

```python
x = data.iloc[:, :-1]
y = data['spices']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**OVR (One-vs-Rest)**




```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class='ovr')
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)*100
```

**Multinomial**

```python
lr1 = LogisticRegression(multi_class='multinomial')
lr1.fit(x_train, y_train)
```

```python
lr1.score(x_test, y_test)*100
```
