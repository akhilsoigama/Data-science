```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_data = [np.random.randint(10,100) for i in range (1000)]
```

```python
pop_table = pd.DataFrame(pop_data, columns=['Population'])
```

```python
np.mean(pop_data)
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x='Population',data=pop_table)
plt.show()
```

```python
import scipy.stats as st
```

```python
s_x = 90
P_u = 82
p_std = 20
n = 81
ap = 0.05
```

```python
z_cal = (s_x-P_u)/(p_std/np.sqrt(n))
z_cal
```

```python
z_table=st.norm.ppf(1-ap)
z_table
```

```python
if z_cal > z_table:
  print('Reject Null Hypothesis')
  print('Accept Alternate Hypothesis')
  print('Mean is not equal to 82')
  print('ha is right')
else:
  print('Accept Null Hypothesis')
  print('reject Alternate Hypothesis')
```
