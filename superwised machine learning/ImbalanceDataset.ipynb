# **Imbalance Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {'Age': np.random.randint(18, 65, 500),
        'EstimatedSalary': np.random.randint(20000, 120000, 500),
        'Purchased': np.random.choice([0, 1], 500, p=[0.7, 0.3])
       }
df = pd.DataFrame(data)
df.head(20)
```

```python
df['Purchased'].value_counts()
```

```python
x =  df.drop(columns=['Purchased'],axis=1)
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
x_rus,y_rus = rus.fit_resample(x,y)
```

```python
y_rus.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_rus,y_rus,test_size=0.2,random_state=42)
```

```python
lr1 = LogisticRegression()
lr1.fit(x_train,y_train)
```

```python
lr1.score(x_test,y_test)*100
```

```python
lr1.predict([[	32,62533	]])
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_ros,y_ros = ros.fit_resample(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_ros,y_ros,test_size=0.2,random_state=42)
```

```python
lr2 = LogisticRegression()
lr2.fit(x_train,y_train)
```

```python
lr2.score(x_test,y_test)*100
```

```python
lr2.predict([[	52,105633	]])
```
