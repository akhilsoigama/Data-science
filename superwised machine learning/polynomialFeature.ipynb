```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

np.random.seed(42)
num_rows = 500

data1 = np.random.randn(num_rows)

a = 0.5
b = 1.0
noise = np.random.randn(num_rows) * 0.5
data2 = a * data1 + b + noise

threshold = 0.5
output = (data1 + data2 > threshold).astype(int)

df = pd.DataFrame({
    'data1': data1,
    'data2': data2,
    'output': output
})
df.head(5)
```

**Without Polynomial feature**

```python
sns.scatterplot(data=df, x='data1', y='data2', hue='output')
plt.show()
```

# **With Polynomial Feature**

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x = pd.DataFrame(poly.fit_transform(x))
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)*100
```

```python
model.predict(x_test)
```
