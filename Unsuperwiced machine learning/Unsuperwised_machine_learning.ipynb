````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# prompt: generate iris data set and column name is sepal_length,sepal_width,petal_lendth,petal_width

from sklearn import datasets
iris = datasets.load_iris()

df = pd.DataFrame(data=np.c_[iris['data']], columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

df.head()
````

````python
sns.pairplot(df)
plt.show()
````

````python
from sklearn.cluster import KMeans
````

````python
wcss = []

for i in range(2,21):
  km = KMeans(n_clusters=i,init= "k-means++")
  km.fit(df)
  wcss.append(km.inertia_)
````

````python
plt.figure(figsize=(12,6))
plt.plot([i for i in range(2,21)],wcss,marker='o')
plt.title("The Elbow Method")
plt.xticks([i for i in range(2,21)])
plt.xlabel("Number of Clusters")
plt.grid(axis='x')
plt.ylabel("WCSS")
plt.show()
````

````python
kmn = KMeans(n_clusters=3)
kmn.fit(df)
df['Predict'] = kmn.predict(df)
````

````python
sns.pairplot(df,hue='Predict')
plt.show()
````
